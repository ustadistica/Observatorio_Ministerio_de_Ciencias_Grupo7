```python
#librerías
import pandas as pd
#Subir archivo
from google.colab import files
#uploaded = files.upload()
# Leer el archivo
df = pd.read_excel("Investigadores_Consolidado (1).xlsx")
# Ver columnas
print(df.columns)
```

```python
# dimensión Universidad
dim_universidad = df[["INST_FILIA"]].dropna().drop_duplicates().reset_index(drop=True)
dim_universidad.insert(0, "Universidad", range(1, len(dim_universidad)+1))

# Ver primeras filas
print(dim_universidad.head(10))
print("Total de universidades:", len(dim_universidad))
```

```python
# dimensión de formación
dim_formacion = df[[
    "ID_NIV_FORMACION_PR",
    "NME_NIV_FORM_PR",
    "NRO_ORDEN_FORM_PR"
]].dropna().drop_duplicates().reset_index(drop=True)

# Ver primeras filas
print(dim_formacion)
print("Total de niveles de formación:", len(dim_formacion))
```

```python
# dimensión residencia
dim_residencia = df[[
    "COD_DANE_RES_PR",
    "NME_PAIS_RES_PR",
    "NME_REGION_RES_PR",
    "NME_DEPARTAMENTO_RES_PR",
    "NME_MUNICIPIO_RES_PR"
]].dropna().drop_duplicates().reset_index(drop=True)

# Ver primeras filas
print(dim_residencia.head(10))
print("Ubicaciones de residencia únicas:", len(dim_residencia))
```

```python
from openpyxl import Workbook

# Crear archivo Excel
with pd.ExcelWriter("dimensiones_investigadores.xlsx", engine="openpyxl") as writer:
    dim_universidad.to_excel(writer, sheet_name="Dim_Universidad", index=False)
    dim_formacion.to_excel(writer, sheet_name="Dim_Formacion", index=False)
    dim_residencia.to_excel(writer, sheet_name="Dim_Residencia", index=False)

files.download("dimensiones_investigadores.xlsx")

```
